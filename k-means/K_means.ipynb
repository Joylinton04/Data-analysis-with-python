```python
## Unsupervised learning
# K-means clustering. K --> represent the number of cluster/groups

# unlabeled dataset into different clusters
# used to organize data into groups based on their similarity
# doesn't require labeled data to work with 

# Algorithms
# choose the number of cluster

## Choosing the right number of cluster(K)
# Elbow method
# Silhouette score. -1 to 1 --> the closer to 1 the better
# No universal Rule
```

```python
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Load and standardize the data
wine = load_wine()
X = wine.data
df = pd.DataFrame(X, columns=wine.feature_names)
df
```

```python
# Standardize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply k means algorithm

from sklearn.cluster import KMeans

kM_model = KMeans(n_clusters=3)
kM_model.fit(X_scaled)
```

```python
labels = kM_model.labels_
centroids = kM_model.cluster_centers_
```

```python
# Visualize 
from sklearn.decomposition import PCA

# Step 1 Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

plt.scatter(pca_result[:,0], pca_result[:,1], c=labels, cmap='viridis', alpha=0.6)
```
