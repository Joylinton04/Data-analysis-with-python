```python
import matplotlib .pyplot as plt
import pandas as pd
```

```python
x_values = ['Python', 'C++','C','Java']
y_values = [25,100,34,56]
```

```python
plt.plot(x_values, y_values, marker='o')
plt.xlabel('Programming Language')
plt.ylabel('Votes')
plt.title('Vote Casting')
plt.xticks(rotation=75)
```

```python
plt.bar(x_values, y_values, alpha=0.9)
plt.xlabel('Programming Language')
plt.ylabel('Votes')
plt.title('Vote Casting')
plt.xticks(rotation=75)

```

```python
plt.pie(y_values, labels=x_values, autopct='%1.0f%%', explode=[0,0.1,0,0])
plt.ylabel('Votes')
plt.title('Vote Casting')
plt.savefig('pie_chart.png', dpi=500)

plt.show()
```
