```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Jack_sales.csv')
df.head()
```

```python
df.tail()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
# scatter plot linear regression accepts data in 2D form
x_values = df[['Ad_Spend_Ghs']] # --> independent variable should be in 2D
y_values = df['Headphone_Sold']

# a model is any algo that we've passed data through it

plt.scatter(x_values, y_values, color="r")
plt.xlabel("Ad Spend Ghs")
plt.ylabel("Headphone sold")
```

```python
model = LinearRegression() # calling upon the model
model.fit(x_values, y_values) # --> training model
# linear regression mostly for forcasting
```

```python
# after training your model
# test the algorithm
# R^2 RMSE MAE

y_predicted = model.predict(x_values) #predicting the y_values
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,root_mean_squared_error
```

```python
r2 = r2_score(y_values, y_predicted) # --> 1 - perfect prediction
print(r2)
mae = mean_absolute_error(y_values, y_predicted) # --> 0 - infinity ->
print(mae)
rms = root_mean_squared_error(y_values, y_predicted) # --> 0 - infinity ->
print(rms)
```

```python
# Actual vs predicted plot
plt.scatter(x_values, y_values, color="blue")
plt.plot(x_values, y_predicted, color="r", alpha=0.8)
plt.xlabel('Ad spend (GHS)')
plt.ylabel('Headphone sold')
```
