```python
# support vector machine (SVM) is a supervised learning algorithm used for both classification and regression tasks in machine learning.
# two lines with a hyperplane between them 
#     /        /      /
#    /maring / margin/
#   /       /       /
#  support vectors --> are critical points tht lie closest to the hyperplane
# the bigger the margin the better the prediction
# Draws a line between two separate variables or data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Jack_sales.csv')
df.head()
```

```python
x_values = df[['Ad_Spend_Ghs']] # --> Independent variable should be in 2D
y_values = df['Headphone_Sold'] # --> Dependent variable should also be in 1D
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split # Splitting of data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.2)
```

```python
svr_model = SVR(kernel='linear', C=1e3, gamma=0.1)
svr_model.fit(x_train, y_train) # --> training process
```

```python
# running test data through the model
y_predict = svr_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
```

```python
r2 = r2_score(y_test, y_predict)
mae = mean_absolute_error(y_test, y_predict)
rmse = root_mean_squared_error(y_test, y_predict)
r2,mae,rmse
```
