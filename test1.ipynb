```python
import pandas as pd
import re
from tabulate import tabulate
```

```python
data = pd.read_csv('FINE GRAINED SENTIMENTS.csv', nrows=10)
data
```

```python
cleaned_characters = data.map(lambda x: re.sub(r'[^a-zA-Z0-9\s.,@!]','',str(x)))
```

```python
cleaned_characters.map(lambda x: x.lower() if isinstance(x,str) else x)
```

```python
with open("cleaned.txt", "w") as f:
    f.write(tabulate(cleaned_characters, headers="keys", tablefmt="psql"))
```
